```python
import numpy as np
import scipy as sp
import qutip as qt
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
def Rx(theta): 
    return (-1j*theta*qt.sigmax()/2).expm()

def Ry(theta): 
    return (-1j*theta*qt.sigmay()/2).expm()

def Rz(phi): 
    return (-1j*phi*qt.sigmaz()/2).expm()

def bloch_vector(qstate):
    rho = qt.ket2dm(qstate)
    return (rho * qt.sigmax()).tr(), (rho * qt.sigmay()).tr(), (rho * qt.sigmaz()).tr()
```

```python
q0 = qt.qstate('d')
q1 = qt.qstate('u')
qp = Ry(np.pi/2) * q0
qm = Ry(3*np.pi/2) * q0
```

```python
bloch_sphere = qt.Bloch()
bloch_sphere.add_vectors([bloch_vector(q0), bloch_vector(qm)])
bloch_sphere
```
